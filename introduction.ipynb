# Overview

A key feature of Markov chains is that the probability of the system being in a certain state depends solely on the previous state of the system. <br> Consider a system which can be in four possible states $A, B, C, D$. If the system is in a certain state at time $t$, there exists a probability that it will transition to another state by time $t+1$. This can be represented in a diagram like the one below:

```python
from IPython.display import Image
Image("markov_diagram.png", width=375, height=350)
```

Let $S_t$ denote the state of the system at time $t$, and let $P(S_t = s)$ denote the probability that the system will be in state $s$ at time $t$. <br> Then we can compute, for example, that if $S_0 = A$ we have:

- $P(S_1 = A) = 0.125$
- $P(S_1 = B) = 0.25$
- $P(S_1 = C) = 0.125$
- $P(S_1 = D) = 0.5$

We can list all the transition probabilities in a matrix like the one below:

```python
from IPython.display import Image
Image("transition_matrix.png", width=375, height=350)
```

Each row contains the transition probabilities of each state. For example, the third row gives us:

- $P(C \rightarrow A) = 0.15$
- $P(C \rightarrow B) = 0.25$
- $P(C \rightarrow C) = 0.30$
- $P(C \rightarrow D) = 0.30$

This will be called the transition matrix, and it will be the basis for many of the functions in this repo.
